```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(x):
  return np.exp(x)/np.sum(np.exp(x),axis=0)
```

```python
x=np.linspace(-10,10)
plt.plot(x,softmax(x))
plt.axis("tight")
plt.title("activation function:softmax")
plt.show()
```
